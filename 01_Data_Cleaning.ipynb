```python
import fitz  # PyMuPDF의 일반적인 임포트

# PDF 파일 열기
doc = fitz.open("석경에이티_분기보고서.pdf")

# 모든 하이라이트된 텍스트를 저장할 변수 초기화
all_highlights = ""

# 모든 페이지를 순회하며 하이라이트된 텍스트 추출
for page in doc:
    # 페이지의 모든 주석(annotations)을 검색
    for annot in page.annots():
        # 주석의 종류가 하이라이트인 경우 처리
        if annot.type[0] == 8:  # 8은 하이라이트를 의미
            rect = annot.rect  # 하이라이트 영역의 좌표
            words = page.get_text("words")  # 페이지의 모든 단어와 그 위치 정보

            # 하이라이트 영역에 포함된 단어 추출
            highlight_text = ""
            for w in words:
                if fitz.Rect(w[:4]).intersects(rect):
                    highlight_text += w[4] + " "

            # 추출된 텍스트를 누적 저장
            all_highlights += highlight_text.strip() + "\n"

# 최종 누적된 하이라이트된 텍스트 출력
print(all_highlights)
```

### 인력 추출

```python
import fitz  # PyMuPDF

# PDF 파일 열기
doc = fitz.open("석경에이티_분기보고서.pdf")

# 결과를 저장할 딕셔너리 초기화
growth_details = {}

# 전체 문서를 순회하며 텍스트 추출
all_text = ""
for page in doc:
    all_text += page.get_text()

# 추출하고자 하는 섹션 찾기
start_phrase = "연구개발인력 현황"
end_phrase = "(3) 주요"

# 시작점과 끝점 찾기
start_index = all_text.find(start_phrase)
end_index = all_text.find(end_phrase)

if start_index != -1 and end_index != -1:
    # 문장 추출
    target_sentence = all_text[start_index:end_index]
    # 문장에서 숫자 추출
    number = ''.join(filter(str.isdigit, target_sentence))

    # 결과를 저장할 딕셔너리 초기화
    extract_dict['총 인력']= f'{number}명'
else:
    extract_dict = "해당 문장을 찾을 수 없습니다."

# 결과 출력
print(extract_dict)
```

### 성장성 추출

```python
import fitz  # PyMuPDF

# PDF 파일 열기
doc = fitz.open("석경에이티_분기보고서.pdf")

# 모든 하이라이트된 텍스트를 저장할 변수 초기화
all_highlights = ""

# 모든 페이지를 순회하며 하이라이트된 텍스트 추출
for page in doc:
    # 페이지의 모든 주석(annotations)을 검색
    for annot in page.annots():
        # 주석의 종류가 하이라이트인 경우 처리
        if annot.type[0] == 8:  # 8은 하이라이트를 의미
            rect = annot.rect  # 하이라이트 영역의 좌표
            words = page.get_text("words")  # 페이지의 모든 단어와 그 위치 정보

            # 하이라이트 영역에 포함된 단어 추출
            highlight_text = ""
            for w in words:
                if fitz.Rect(w[:4]).intersects(rect):
                    highlight_text += w[4] + " "

            # 추출된 텍스트를 누적 저장
            all_highlights += highlight_text.strip() + "\n"

# 추출하고자 하는 섹션 찾기
start_phrase = "마. 시장의 주요 특성ㆍ규모 및 성장성"
end_phrase = ")규모입니다."

# 시작점과 끝점 찾기
start_index = all_highlights.find(start_phrase)
end_index = all_highlights.find(end_phrase)

if start_index != -1 and end_index != -1:
    # 시작점과 끝점 사이의 텍스트 추출
    extracted_text = all_highlights[start_index:end_index + len(end_phrase)]
    # 딕셔너리에 저장
    extract_dict['성장성'] = extracted_text
else:
    print("해당 텍스트를 찾을 수 없습니다.")

print(extract_dict)
```

```python
from openai import OpenAI # openai==1.2.0

client = OpenAI(
  api_key="up_L4wu6Az9MjVfUQH7FLEUwozfvfJZa",
  base_url="https://api.upstage.ai/v1/solar"
)

# this system prompt will be used globally on the chat session.
system_prompt = {
    "role": "system",
    "content": "You are a helpful assistant."
}

model = "solar-1-mini-chat"

# Keep chat history in a list.
chat_history = []

# Set a limit for the chat history to manage the token count.
history_size = 10

while True:
    # Step 1: Get user input.
    user_prompt = input("User: ")
    chat_history.append({
        "role": "user",
        "content": user_prompt
    })

    # Step 2: Use the chat history to generate the chatbot's response.
    messages = [system_prompt] + chat_history
    stream = client.chat.completions.create(
        model=model,
        messages=messages,
        stream=True,
    )

    # Step 3: Output the Solar response chunks.
    print("SOLAR: ", end="")
    solar_response = ""
    for chunk in stream:
        response_content = chunk.choices[0].delta.content
        if response_content is not None:
            solar_response += response_content
            print(response_content, end="")
    print() # Finish response output.

    # Step 4: Append the Solar response to the chat history.
    chat_history.append({
        "role": "assistant",
        "content": solar_response
    })

    # Step 5: Ensure the chat history doesn't exceed the size limit.
    chat_history = chat_history[:history_size]
```
